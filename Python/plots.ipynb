```python
ages=np.random.normal(20,1.5,1000)
bin = [ages.min(),ages.max()]
for  i in range(int(ages.min()),int(ages.max()),2):
    bin.append(i)
bin.sort()  # Fix: Add parentheses to actually call the sort() method
```

```python
plt.hist(ages,bins=bin)
plt.show()
```
